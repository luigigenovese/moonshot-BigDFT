```python
from numpy import loadtxt
pockets = {}
for res,pkt in loadtxt('pockets.txt',dtype=str,usecols=[0,1]):
    pockets.setdefault(pkt,[]).append(res)
chains = [{'title': 'pockets', 'cats': pockets}]
```
